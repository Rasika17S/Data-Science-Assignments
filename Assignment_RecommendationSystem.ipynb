```python
# importing libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# loading dataset
df = pd.read_csv(r"C:\Users\rasika\Downloads\anime.csv")
df.head()
```

```python
# Dataset Summary
df.info() # numbers of entries, number of columns and there datatypes
```

```python
# Summary statistics
df.describe()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# replacing missing values with mode
df['genre'] = df['genre'].fillna(df['genre'].mode()[0])
```

```python
# replacing missing values with mode
df['type'] = df['type'].fillna(df['type'].mode()[0])
df['type'].dtype
```

```python
# replacing missing values with mean
df['rating'] = df['rating'].fillna(df['rating'].mean())
```

#### Feature Encoding

```python
le = LabelEncoder()
df['type'] = le.fit_transform(df['type'])
print(df.head())
print(df['type'].dtype)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
genre_matrix = tfidf.fit_transform(df['genre'])
genre_matrix
```

```python
# feature Extraction
tfidf.get_feature_names_out()
```

```python
tfidf_df = pd.DataFrame(genre_matrix.toarray(), columns=tfidf.get_feature_names_out())
tfidf_df.head()
```

```python
#Converting object column to nuimeric type
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')

#replacing missing('unknown') with median
df['episodes'] = df['episodes'].fillna(df['episodes'].median())
```

```python
# Minmaxscaling
mm = MinMaxScaler()
num_column = df.drop('anime_id', axis = 1).select_dtypes(include = ['int', 'float']).columns
df[num_column] = mm.fit_transform(df[num_column])
df.head()
```

```python
# Combining all features
from scipy.sparse import hstack
feature_matrix = hstack([genre_matrix, df[['episodes']]])
feature_matrix
```

```python
# Calculating cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(feature_matrix)
```

```python
# Index mapping
indices = pd.Series(df.index, index=df['name']).drop_duplicates()
```

```python
idx = indices['Naruto']
idx
```

#### Recommendation Function

```python
def recommend_anime(title, top_n=10):

    idx = indices[title]

    similarity_scores = list(enumerate(cosine_sim[idx]))

    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    similarity_scores = similarity_scores[1:top_n+1]

    anime_indices = [i[0] for i in similarity_scores]

    return df[['name', 'genre','rating']].iloc[anime_indices]
```

```python
# Adding threshold
def recommend_with_threshold(
        title,
        threshold=0.5):

    idx = indices[title]

    similarity_scores = list(enumerate(cosine_sim[idx]))

    filtered_scores = []

    for anime_idx, score in similarity_scores:

        if score > threshold:
            filtered_scores.append((anime_idx, score))

    filtered_scores = sorted(filtered_scores, key=lambda x: x[1], reverse=True)

    filtered_scores = filtered_scores[1:]

    anime_indices = [i[0] for i in filtered_scores]

    return df[['name', 'genre','rating']].iloc[anime_indices]
```

```python
# Experimenting with threshold with 0.3
recommend_with_threshold("Naruto", threshold=0.3)
```

```python
# Experiment with treshold 0.5
recommend_with_threshold("Naruto", threshold=0.5)
```

```python
# Experiment with threshold 0.7
recommend_with_threshold("Naruto", threshold=0.7)
```

```python
# Comparing thresholds
for t in [0.3, 0.5, 0.7]:

    result = recommend_with_threshold("Naruto", threshold=t)

    print("Threshold =", t)
    print("Number of Recommendations =", len(result))
    print("-"*50)
```

#### Performance Analysis
1. At threshold 0.3, the system generated a large number of recommendations. Some recommendations were less relevant because the similarity requirement was low.
2. At threshold 0.5, the recommendations were more balanced and relevant.
3. At threshold 0.7, only highly similar anime were recommended, resulting in fewer but more accurate suggestions.

#### Areas of improovemnt
1. Current system uses anime features only. Using user-watch history would improve personalization.
2. Current approach is content-based. Collaborative filtering can capture user preferences better.

#### Interview Questions

1. Difference between User-Based and Item-Based Collaborative Filtering

User-Based Collaborative Filtering
"Users similar to me liked these items."

Example:
           
            Rasika likes:
            
            Naruto
            
            One Piece

User B likes:

            Naruto
          
            One Piece
          
            Bleach

Recommend:
           
            Bleach

Because both users have similar preferences.

Item-Based Collaborative Filtering
"Items similar to the item you liked."

Example:

            Naruto is similar to:

            Bleach
            
            One Piece

Recommend:
            
            Bleach
            
            One Piece

2. What is Collaborative Filtering?

Collaborative filtering recommends items based on user behavior.

It assumes, Users with similar interests will like similar items.
