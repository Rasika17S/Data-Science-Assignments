# Descriptive Analytics for Numerical Columns

```python
# Loading the dataset 
df = pd.read_csv(r"C:\Users\rasika\Downloads\sales_data_with_discounts (1).csv")
df
```

```python
df.describe()
```

```python
# Identify numerical columns in the dataset.
numerical_columns = df.select_dtypes(include = ['int', 'float']).columns
numerical_columns
```

```python
# Calculate the mean, median, mode, and standard deviation for these columns.
# Mean - represented by sum of all the values divided by total number of values.
# In following code lines it will give average of all the values of particular column

# Median - It gives middle value from all the values, if there are even number of values it will give average of middle two values as median.

# Mode - It gives value that is most frequantly occured. It can be Bimodal(2 values occured frequantly), 
# trimodal(3 values occured frequantly), multimodal(more that 3 values occured frequantly).

# Standard Deviation - It gives measure of how data values are spreded close to average(mean).
# values closer to mean  - low standard deviation, values far from mean high standard deviation

print("Mean, Median and Standard Deviation of Volume")
print(df['Volume'].mean())
print(df['Volume'].median())
print(df['Volume'].std())
print('\n')
print("Mean, median and Standard Deviation of Average Price")
print(df['Avg Price'].mean())
print(df['Avg Price'].median())
print(df['Avg Price'].std())
print('\n')
print("Mean, median and Standard Deviation of Total Sales Values")
print(df['Total Sales Value'].mean())
print(df['Total Sales Value'].median())
print(df['Total Sales Value'].std())
print('\n')
print("Mean, median and Standard Deviation of Discount Rate")
print(df['Discount Rate (%)'].mean())
print(df['Discount Rate (%)'].median())
print(df['Discount Rate (%)'].std())
print('\n')
print("Mean, median and Standard Deviation of Discount Amount")
print(df['Discount Amount'].mean())
print(df['Discount Amount'].median())
print(df['Discount Amount'].std())
print('\n')
print("Mean, median and Standard Deviation of Net Sales Values")
print(df['Net Sales Value'].mean())
print(df['Net Sales Value'].median())
print(df['Net Sales Value'].std())
```
