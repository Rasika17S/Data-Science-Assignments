```python
# 1. Create a list of 5 integers. 
# Perform and print the result of the following operations: 
        # append, extend, insert, remove, pop, clear, index, count, sort, and reverse.
```

```python
lst = [10, 20, 30, 40, 50]
# append
lst.append(60)
print(lst)
# insert
lst.insert(2, 70)
print(lst)
# remove
lst.remove(70)
print(lst)
# pop
lst.pop()
print(lst)
# clear
lst.clear()
print(lst)
```

```python
# extend
a = [10, 20]
b = [30, 40]
a.extend(b)
print(a)
```

```python
lst1 = [1, 2, 3, 4, 5, 2, 3, 4, 2]
# index
lst1.index(2) # it will give index position of first appearance of 2 

# indexing
print(lst1[0]) # will give value at index position 0
print(lst1[:]) # will give entire list
print(lst1[::-1]) # will give list in reverse order
print(lst1[-1]) # will give value at last position
# Slicing 
print(lst1[0:5]) # fetch value from index 0 to 4
print(lst1[:-1]) # fetch values from first index posiion till index -2 
```

```python
# count
lst1.count(2)
```

```python
# sort
lst1.sort()
lst1
```

```python
# reverse
lst1.reverse()
lst1
```

```python
'''6. Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail' '''

marks = int(input('Enter your marks'))
if marks >= 90:                                       # 90 and above: 'A'
    print('You have passed the exam with grade A')
elif (marks >= 75) and (marks <= 89):                 # 75–89: 'B'
    print('You have passed the exam with grade B')
elif marks >= 50 and marks <= 74:                     # 50–74: 'C'
    print('You have passed the exam with grade C')
else:                                                 # Below 50: 'Fail
    print("You have failed the exam")
```

```python
# 8. Create:
'''- A scalar using np.array(5)
   - A 1D array with values 1 to 5
   - A 2D array (2x3) with values from 10 to 60 in steps of 10'''

import numpy as np

scalar = np.array(5)  # A scalar using np.array(5)
print(scalar)

array1 = np.array([1, 2, 3, 4, 5]) # A 1D array with values 1 to 5
print(array1, array1.ndim)

array2 = np.arange(10, 70, 10).reshape(2, 3) # A 2D array (2x3) with values from 10 to 60 in steps of 10
print(array2)
```

## Section C: Exploring Pandas

```python
# 11.Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
''''- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows'''
```

```python
import pandas as pd
data = {
    'Name': ['Raj', 'Vishal', 'Shreya', 'Priya', 'Sagar', 'Anna', 'Vishal', 'Ben', 'Cary', 'John'],
    'Age': [23, 25, 33, 45, 22, 30, 25, 50, 27, 28],
    'City': ['Mumbai', 'Pune', 'Belgaum', 'Bengalore', 'Hydrabad', 'Singapore', 'Pune', 'Shimla', 'America', 'California'],
    'Salary': [25000, 50000, 35000, np.nan, 30000, 60000, 50000, 20000, 30000, 80000]
}

df = pd.DataFrame(data)
df
```

```python
# info
df.info()
```

```python
# decribe
df.describe()
```

```python
# Select 'Name' and 'City' columns
df[['Name', 'Age']]
```

```python
# Drop 'City' column

df.drop('City', axis = 1)
```

```python
# Check for duplicates

print(df.duplicated().sum())
df.duplicated()
```

```python
# Remove  duplicate row for data of vishal using drop_duplicates
df.drop_duplicates().reset_index()
```

```python
# Checking for missing value
df.isnull().sum()
```

```python
# calculating mean
salary = df['Salary'].mean()
salary
```

```python
# Fill  missing values in 'Salary' column with the mean
df.fillna(salary)
```
