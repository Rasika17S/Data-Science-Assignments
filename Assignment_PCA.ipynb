```python
X = df.drop('Type', axis = 1)
y = df['Type']
```

```python
# Standardization
sc = StandardScaler()
X_sc = sc.fit_transform(X)
X_sc
```

```python
# Fitting PCA using all components
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(X_sc)
```

### Finding Optimal numbers of PC's

#### Scree Plot
#### Explains Variance ratio
#### Shows Elbow Methoed

```python
plt.plot(
    range(1, 14),
    pca.explained_variance_ratio_,
    marker = 'o')
```

#### Cumulative Variance

```python
plt.plot(range(1, 14),
         np.cumsum(pca.explained_variance_ratio_),
         marker = 'o')
```

### The scree plot and cumsum shows a clear elbow at the 4th principal component, indicating that the first four components capture most of the important information in the dataset.

```python
# Fiting PCA on first 4 features
pca = PCA(n_components = 4)
X_pca_final = pca.fit_transform(X_sc)
X_pca_final
```

### Clustering On Original Dataset

```python
# Finding best k
from sklearn.cluster import KMeans

wcss = []

for k in range(2, 11):
    kmeans = KMeans(
        n_clusters = k,
        random_state = 42)
    kmeans.fit_transform(X_sc)
    wcss.append(kmeans.inertia_)
```

```python
# Elbow method to find best k
plt.plot(range(2, 11),
         wcss,
         marker = 'o'
)
```

```python
# Applying kmeans on k  = 3 clusters
kmeans_original = KMeans(
    n_clusters = 3,
    random_state = 42
)
cluster_original = kmeans_original.fit_predict(X_sc)
```

```python
### Evaluation with Silhouette Score
from sklearn.metrics import silhouette_score

sil_original = silhouette_score(
    X_sc,
    cluster_original)
sil_original
```

```python
pca_vis = PCA(n_components=2)

X_vis = pca_vis.fit_transform(X_sc)

plt.scatter(
    X_vis[:,0],
    X_vis[:,1],
    c=cluster_original
)
```

### Clustering on PCA

```python
wcss = []

for k in range(2, 11):
    kmeans = KMeans(
        n_clusters = k,
        random_state = 42)
    kmeans.fit_transform(X_pca_final)
    wcss.append(kmeans.inertia_)
```

```python
# Elbow method to find best k
plt.plot(range(2, 11),
         wcss,
         marker = 'o'
)
```

```python
# Applying kmeans on k  = 3 clusters
kmeans_original = KMeans(
    n_clusters = 3,
    random_state = 42
)
cluster_pca = kmeans_original.fit_predict(X_pca_final)
```

```python
### Evaluation with Silhouette Score
sil_pca = silhouette_score(
    X_pca_final,
    cluster_pca)
sil_pca
```

```python
# Visualization
plt.scatter(
    X_pca_final[:,0],
    X_pca_final[:,1],
    c=cluster_pca
)
```

### Comparision Analysis

```python
print('Silhouette Score of Original K-Means', sil_original)
print('Silhouette Score of K-Means with PCA', sil_pca)
```

### K-Means clustering on PCA-transformed data produced better cluster quality than K-Means on the original feature space.

##### Similarities
1. Same number of clusters
2. Similar cluster patterns

##### Differences
1. PCA may produce cleaner clusters
2. Some information may be lost

##### Impact of dimensionality reduction on cluster 
1. Reduced Noise
2. Improved clustering performance

#### Trade-off's
1. clustering on original dataset,  more features more noise.
2. Clustering on PCA transformed dataset, some information will be lost.

### Conluclusion and insights
#### Key Findings
1. EDA helped understand distributions, outliers, and feature relationships.
2. PCA reduced dimensionality while preserving most information.
3. K-Means clustering was applied on both original and PCA-transformed datasets.
4. PCA-based clustering showed better separation of clusters and improved evaluation metrics.

#### Practical Implications
1. PCA is useful when datasets contain many correlated features.
2. Clustering helps discover hidden groups in data without labels.
3. Combining PCA with clustering often improves performance and visualization.

#### Recommendations
1. Use original data when feature interpretation is important.
2. Use PCA when the dataset has many features, high correlation, or noise.
3. For large datasets, PCA + K-Means is usually more efficient and often produces better clustering results.
