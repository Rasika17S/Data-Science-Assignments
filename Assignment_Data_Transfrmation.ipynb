### One-Hot Encoding
Pros
Does not create any order or ranking among categories.
1. Suitable for nominal categorical data such as gender in this dataset.
2. Prevents the model from assuming one category is greater than another.

Cons
1. Increases the number of columns in the dataset.
2. Can lead to high dimensionality when a feature has many unique categories.
3. Requires more memory and computational resources.
Example:
Sex - 
Male,
Female

### Label Encoding
Pros
1. Simple and easy to implement.
2. Does not increase the number of columns.

Cons
1. Introduces numerical labels that may imply an order where none exists.
2. Different labels may be interpreted as having greater or lesser importance.

### Feature Engineering

```python
# Net Capital - 
# The dataset contains two financial columns:  capital_gain, capital_loss
# Instead of treating them separately, we can create a new feature representing the overall capital position of an individual.
# Combines gains and losses into a single meaningful feature.
# Represents the individual's net financial benefit.
# May have a stronger relationship with income than the original features separately.

df['net_capital'] = df['capital_gain'] - df['capital_loss']
df[['capital_gain', 'capital_loss', 'net_capital']].head()
```

```python
# Working Age Group
# People in different age groups often have different earning capacities and work experience levels.
# Groups similar ages together.
# Income often varies across age groups due to experience and seniority.

df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 25, 45, 65, 100],
    labels=['Young', 'Adult', 'Middle_Aged', 'Senior']
)
df[['age', 'age_group']].head()
```
