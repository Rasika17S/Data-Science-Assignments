###  filling missing values with mean of entire column

```python
df['LB'] = df['LB'].fillna(df['LB'].mean())
df['AC'] = df['AC'].fillna(df['AC'].mean())
df['DS'] = df['DS'].fillna(df['DS'].mean())
df['DP'] = df['DP'].fillna(df['DP'].mean())
df['MLTV'] = df['MLTV'].fillna(df['MLTV'].mean())
df['Width'] = df['Width'].fillna(df['Width'].mean())
df['Tendency'] = df['Tendency'].fillna(df['Tendency'].mean())
df['NSP'] = df['NSP'].fillna(df['NSP'].mean())
```

```python
# after handling missing values no column contain missing value.
df.isnull().sum()
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# removing duplicates
df.drop_duplicates(inplace = True)
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# Identify and correct any inconsistencies in data types (e.g., numerical values stored as strings).
# No such type of row appeared in dataset
```

## Detect And Treat Outliers

```python
df['LB'].plot(kind = 'box')
```

```python
Q1 = df['LB'].quantile(0.25)
Q3 = df['LB'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['LB'] >= lower) & (df['LB'] <= upper)].copy()
```

```python
df['LB'].plot(kind = 'box')
```

```python
df['AC'].plot(kind = 'box')
```

```python
Q1 = df['AC'].quantile(0.25)
Q3 = df['AC'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['AC'] = df['AC'].clip(lower, upper)
```

## Footfall Over Month using Footfall dataset

```python
# reading FootFall dataset
foot_fall = pd.read_csv(r"C:\Users\rasika\Downloads\footfalls.csv")
foot_fall
```

```python
# converting "Month" column to datetime datatype
foot_fall['Month'] = pd.to_datetime(foot_fall['Month'], format = '%b-%y')
foot_fall['Month'].head()
```

```python
# Setting Month as index
foot_fall.set_index('Month', inplace = True)
```

```python
foot_fall.head()
```

```python
# linechart to show trend over time that is 'FootFall' over 'Month'
plt.figure(figsize = (10, 5))
plt.plot(foot_fall, marker = 'o')
plt.title('FootFall over Month')
plt.xlabel('Month')
plt.ylabel('FootFall')
plt.show()
```

The exploratory analysis of the cardiographic dataset revealed important characteristics of fetal heart monitoring parameters. Missing values and duplicate records were handled appropriately, and outliers were identified using boxplot and removed using the IQR method.

In Histogram, data is normally distributed in LB, MLTV, Width and ASTV. Width and ASTV are slightly in bell shaped curve form.
Most of the varibales are Positively skewed and contains outliers.

Pairplot, scatterplot visulaizations show that no features have relation with other feature.

The FootFall dataset demonstrates a clear increasing trend over the observed period, indicating that the number of visitors has generally grown from 1991 to 2004. The repeated peaks and troughs suggest the presence of seasonal effects, while the frequent fluctuations indicate the influence of short-term factors on customer visits.
The highest footfall values are recorded during the final years of the dataset, showing improved performance and increased customer engagement over time.
