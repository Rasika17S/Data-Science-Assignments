## Hospital Patient Data Analysis

```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# Loading the patient dataset 
patient_data = pd.read_csv(r"C:\Users\rasika\Downloads\Patient_Data (1).csv")
df = pd.DataFrame(patient_data)
df
```

```python
# Summary of the dataset
df.info()
```

```python
# Droping administrative columns like ['ReceptionistID', 'CheckInTime'].
df.drop(['ReceptionistID', 'CheckInTime'], axis = 1, inplace = True)
df
```

```python
# Use groupby to find total bill amount per department.
df.groupby('Department')['BillAmount'].sum()
```

```python
# Remove duplicate patient records based on PatientID.
df.drop_duplicates(inplace = True)
df
```

```python
# Fill missing BillAmount values with the mean bill amount.
BillAmount = df['BillAmount'].mean()
df.fillna(BillAmount, inplace = True)
```

```python
Billing_data = pd.read_csv(r"C:\Users\rasika\Downloads\Billing_Data (1).csv")
Billing_data
```

```python
# Merge the billing dataset with patient dataset on PatientID.
df1 = pd.merge(df, Billing_data, on = 'PatientID', how = 'inner')
df1
```

```python
# Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
new_patients = pd.DataFrame({
    'PatientID' : [106, 107, 108, 109, 110],
    'Name' : ['Methew', 'John', 'Harry', 'Nickol', 'Nick'],
    'Department' : ['Cardiology', 'Neurology', 'Orthopedics', 'Dermatology', 'Cardiology'],
    'Doctor' : ['Dr.Jonas', 'Dr.Jen', 'Dr.Lee', 'Dr.Kevin', 'Dr.John'],
    'BillAmount' : [6000, 4520, 10000, 7000, 2000],
})
```

```python
new_patients
```

```python
current_week = pd.concat([df, new_patients], ignore_index = True)
current_week
```

```python
# Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).# 
new_bills = pd.DataFrame({
    'InsuranceCovered' : [3000, 1500, 4000, 2000, 500],
    'FinalAmount' : [3000, 3020, 6000, 5000, 1500]
})
new_bills
```

```python
updated_info = pd.concat([new_patients,new_bills], axis = 1)
updated_info
```
