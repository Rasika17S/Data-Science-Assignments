```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset
df = pd.read_csv(r"C:\Users\rasika\Downloads\sonardataset.csv")
df.head()
```

```python
# Feature encoding on target column 'Y'
y = df['Y']

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

y_le = le.fit_transform(y)
```

```python
# Feature scaling to adjust features in one range
X = df.drop('Y', axis = 1)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

```python
# Train test split with 80% data for training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sc, y_le, test_size = 0.2, random_state = 42, stratify = y)
```

### BUilding ANN model

```python
!pip install tensorflow
```

```python
from tensorflow.keras.models import Sequential # Creates neural network where layers are added one after another
from tensorflow.keras.layers import Dense # creates fully connected layer

model = Sequential() # creates empty neura network

model.add(Dense(32, activation = 'sigmoid', input_shape = (60, ))) # Creates hidden layer with 32 neurons, uses sigmoid function cause it is binary classification type of dataset and takes all 60 input features
model.add(Dense(16, activation = 'sigmoid')) # creates second hidden layer with 16 neurons and forms more complex pattern
model.add(Dense(1, activation = 'sigmoid')) # creates output layer with sigmoid that will range output between 0 and 1
```

### Compiling model

```python
model.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy']) 
# tells tensorflow how neural network should run.
# optimizer controls the weight updates and biases to reduce the errors
# loss tells how wrong predictions are 
# metrics to evalaute performance
```

### Training the model

```python
history = model.fit(X_train, y_train, epochs = 100, batch_size = 16, validation_split = 0.2, verbose = 1)
```

### Predictions

```python
y_prob = model.predict(X_test)
y_pred = (y_prob > 0.5).astype(int)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy Score ", accuracy)
print("precision Score ", precision)
print("Recall Score ", recall)
print("F1 Score ", f1)
```

### HpyerParametr Tunning using grid search

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def create_model(neurons=32):

    model = Sequential()

    model.add(Dense(neurons,
                    activation='relu',
                    input_shape=(60,)))

    model.add(Dense(1,
                    activation='sigmoid'))

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
model = KerasClassifier(
    model=create_model,
    verbose=0
)
```

```python
param_grid = {
    "model__neurons":[64],
    "batch_size":[32],
    "epochs":[100]
}

grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy'
)

grid.fit(X_train, y_train)
```

```python
# Best parameters and score
print(grid.best_params_)
print(grid.best_score_) 
```

### Metrics Evaluation on tunned model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
#getting best model
best_model = grid.best_estimator_

# getting prediction with best model
y_prob_best = best_model.predict(X_test)
y_pred_best = (y_prob_best > 0.5).astype(int)

# Best Model evaluation
accuracy_grid = accuracy_score(y_test, y_pred_best)
precision_grid = precision_score(y_test, y_pred_best)
recall_grid = recall_score(y_test, y_pred_best)
f1_grid = f1_score(y_test, y_pred_best)

print("Accuracy Score ", accuracy_grid)
print("precision Score ", precision_grid)
print("Recall Score ", recall_grid)
print("F1 Score ", f1_grid)
```
